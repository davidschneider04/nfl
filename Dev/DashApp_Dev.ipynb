```python
def build_datadict(team):
    df = nyd_team[nyd_team['posteam']==team]
    xvals = list(df['ydsnet'])
    yvals = list(df['num_plays'])
    datadict = {'x': xvals
               ,'y': yvals
               ,'type': 'bar'
               ,'name': team}
    b = go.Bar(name=team, x=xvals, y=yvals)
    return b

ddict_list = [build_datadict(team) for team in nyd_team['posteam'].unique()]
print(ddict_list)
```

```python
app.layout = html.Div(children=[
    html.H1(children="2018 NFL Play Breakdown")
    ,dcc.Graph(id='graph'
               ,figure={'data': ddict_list
                        ,'layout': {'plot_bgcolor': colors['text']
                                    ,'paper_bgcolor': colors['text']
                                    ,'title': "Dash Data Visualization"}
                       }
     
              )
    ,dcc.RangeSlider(id='year-range-slider'
               ,min=net_yard['ydsnet'].min()
               ,max=net_yard['ydsnet'].max()
               ,step=1
               ,marks={i: '{}'.format(i) for i in range(net_yard['ydsnet'].min(), net_yard['ydsnet'].max(),5)}
               ,value=[net_yard['ydsnet'].min(), net_yard['ydsnet'].max()])
])
```
